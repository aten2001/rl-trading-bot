```python
prices = get_data(['JPM'], pd.date_range(start_date, end_date))
prices.fillna(method='ffill', inplace=True)
prices.fillna(method='bfill', inplace=True)
prices['Cash'] = 1.0
positions = pd.DataFrame().reindex_like(prices)
positions.fillna(0, inplace=True)
positions.iloc[0, -1] = 100000
```

```python
positions
```

```python
for date, row in orders.iloc[:-1,:].iterrows():
    if date not in positions.index:
        continue
    symbol = row['Symbol']
    shares = row['Shares']
    action = row['Order']
    price = prices.loc[date, symbol]
    if action == 'SELL':
        shares = shares * -1
    positions.loc[date, symbol] += shares
    cash_decrease = shares * price
    if cash_decrease > 0:
        cash_decrease *= (1 + 0)
    elif action == 'BUY':
        cash_decrease *= (1 - 0)
    cash_decrease += 0
    positions.loc[date, 'Cash'] -= cash_decrease
positions = positions.cumsum(axis=0)
portval = (positions * prices).sum(axis=1)
portval = portval.to_frame('PortVal')
```
