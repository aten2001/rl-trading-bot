```python
import pandas as pd
import numpy as np
import datetime as dt
import os
import sys
sys.path.insert(0, './')
import util
from collections import OrderedDict
from datetime import datetime
```

```python
orders = pd.read_csv("./orders/orders-01.csv", header=0, parse_dates=True, na_values=['nan'])
orders.sort_values(by='Date', inplace=True)
start_date = orders.loc[0, 'Date']
end_date = orders.iloc[-1, 0]
orders.set_index('Date', inplace=True)
orders
```

```python
pd.date_range(start_date, end_date)
stocks = list(orders.Symbol.unique())
prices = util.get_data(stocks, pd.date_range(start_date, end_date))
prices.fillna(method='ffill', inplace=True)
prices.fillna(method='bfill', inplace=True)
prices['Cash'] = 1.0
prices
```

```python
positions = pd.DataFrame().reindex_like(prices)
positions.fillna(0, inplace=True)
positions.iloc[0, -1] = 1000000
for date, row in orders.iterrows():
    if date not in positions.index:
        continue
    symbol = row['Symbol']
    shares = row['Shares']
    action = row['Order']
    price = prices.loc[date, symbol]
    if action == 'SELL':
        shares = shares * -1
    positions.loc[date, symbol] += shares
    positions.loc[date, 'Cash'] -= shares * price
positions = positions.cumsum(axis=0)
positions
```

```python
portval = (positions * prices).sum(axis=1)
portval.to_frame()
```

```python
headers = stocks + ['Cash', 'Date', 'Port_Val']
date_format = "%Y-%m-%d"
a = datetime.strptime(start_date, date_format)
b = datetime.strptime(end_date, date_format)
delta = b - a
df_trades = pd.DataFrame(0, index=range(delta.days+1), columns=headers)
df_trades['Date'] = pd.date_range(start=start_date, end=end_date)
df_trades = df_trades.set_index('Date')
df_trades.loc[start_date, 'Cash'] = 1000000
```

```python
for index, row in orders.iterrows():
    date = row['Date']
    stock = row['Symbol']
    shares = row['Shares']
    action = row['Order']
    if action == 'SELL':
        shares = shares * -1
    df_trades.loc[date, stock] += shares
df_holdings = df_trades.copy(deep=True)
for stock in list(df_trades.axes[1])[:-2]:
    df_holdings[stock] = df_trades[stock].cumsum()
```
